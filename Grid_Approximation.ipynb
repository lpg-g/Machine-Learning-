```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import expon

random_values = norm.rvs(size=100, loc=0.5, scale=0.5)
mus = np.linspace(0, 3, num=100)
sigmas = np.linspace(0, 1, num=100)

x = []
y = []
p = []

for mu in mus:
    for sigma in sigmas:
        posterior = np.prod(norm.pdf(x=random_values, loc=mu, scale=sigma)) * \
                    norm.pdf(x=mu, loc=0, scale=3) * expon.pdf(x=sigma, scale=1)
        x.append(mu)
        y.append(sigma)
        p.append(posterior)

plt.scatter(x, y, c=p)
plt.show()
```
